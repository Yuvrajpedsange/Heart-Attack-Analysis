```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart=pd.read_csv("Heart.csv")
```

```python
heart.head()
```

```python
heart.shape
```

```python
heart.info()
```

```python
heart.rename(columns={'cp':'Chest_pain_type','trtbps':'Resting_blood_pressure','chol':'cholestoral','fbs':'Fasting_Blood_sugar','restecg':'resting_electrocardiographic','thalachh':'Max_heart_rate','exng':'exercise_induced_angina','oldpeak':'Previous_peak','slp':'Slope','caa':'major_vessels_impacted','thall':'Thal_rate','output':'Heart_attack_chances'},inplace=True)
```

```python
heart.info()
```

```python
heart[['sex','Chest_pain_type','Fasting_Blood_sugar','resting_electrocardiographic','exercise_induced_angina','Heart_attack_chances']]=heart[['sex','Chest_pain_type','Fasting_Blood_sugar','resting_electrocardiographic','exercise_induced_angina','Heart_attack_chances']].astype(str)
```

```python
heart.info()
```

```python
heart['sex'].replace({'1':'Male','0':'Female'},inplace=True)
```

```python
heart.head()
```

```python
heart['Chest_pain_type'].replace({'0':'No pain','1':'typical angina','2':'atypical angina','3':'non-anginal pain','4':'asymptomatic'},inplace=True)
```

```python
heart['Fasting_Blood_sugar'].replace({'1':'Greater than 120','0':'Less than 120'},inplace=True)
```

```python
heart['resting_electrocardiographic'].replace({'0':'normal','1':'abnormal','2':'left pump chamber wont work properly'},inplace=True)
```

```python
heart['exercise_induced_angina'].replace({'1':'True','0':'False'},inplace=True)
```

```python
heart['Heart_attack_chances'].replace({'1':'More chances','0':'Less chances'},inplace=True)
```

```python
heart=pd.DataFrame(heart)
condition=(heart['Heart_attack_chances']=='Less chances')
heart.drop(heart[condition].index,inplace=True)
```

```python
heart.head()
```

```python
heart.shape
```
